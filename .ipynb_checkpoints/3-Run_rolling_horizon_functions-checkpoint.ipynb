# Call rolling horizon functions

```python
from scen_generation_new_generate import generateForecast
import numpy as np
import time

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.colors import LightSource
#%matplotlib notebook
```

```python
from upside import runDP, plotVF, buildOfferCurves, runThreeStage, runThreeStageBorder, plotOfferCurves
#from upside import offer_curve_as_fcn
```

Load data:

```python
which_day = 25 
num_scenarios = 4
Bm_dict1, Prob_ay1 = generateForecast(which_day, num_scenarios)
Bm_dict2, _ = generateForecast(which_day+1, num_scenarios)
```

```python
# Set parameters:
PARAMS = {
    'RES'   : 10,        # resolution of discretization (10)
    'MDP'   : -10**5,    # penalty in DP model
    'MST'   : 10**5,     # penalty in stochastic model
    'QDA'   : 30,        # energy contracted in the DA (30)
    'C0'    :  0,        # thermal unit fixed op. cost
    'MC'    : 25,        # diesel marginal cost (25)
    'CST'   :  5,        # diesel start-up cost
    'CSH'   :  5,        # diesel shut-down cost
    'LMIN'  :  0,        # storage min. level
    'LMAX'  : 80,        # storage max. level (80)
    'DMIN'  :  0,        # diesel min. output
    'DMAX'  : 70,        # diesel max. output (70)
    'DRAMP' : 30,        # diesel max ramp up/dw (70)
    'PUP'   : 30,        # max. battery charge (30)
    'PDW'   : 30,        # max. battery discharge (30)
    'L0'    : 30,        # storage level in start of t=0 (3-stage model) (30)
    'D0'    : 30,        # diesel status in start of t=0 (3-stage model) (30)
    'U0'    :  1,        # uc status of diesel in start of t=0 (3-stage model)
    'ETA'   :  1,        # storage charging efficiency
    'VFfac' : 0.25,       # Value function factor in objective
}
```

Look-ahead version:

```python
%matplotlib inline
useVF=True
log=True

offcurves = buildOfferCurves()
L0 = PARAMS['L0']
D0 = PARAMS['D0']
profitAll = []
penalAll = np.array([0,0,0,0])

for h in range(1,25):
    if h <= 21:
        # DP
        CC, FF = runDP(hour=h, params=PARAMS, data=Bm_dict1, data2=Bm_dict2, probMat=Prob_ay1)

        # Three-stage 
        oc_new, newL0, newD0, profit_new, profit0, penal = runThreeStage(hour=h, params=PARAMS, data=Bm_dict1, probMat=Prob_ay1, 
                                                           offerCurves=offcurves, L0=L0,D0=D0, CC=CC, useVF=useVF, log=log)

        # Update offer curves
        offcurvesAux = {}
        offcurvesAux[0] = offcurves[1]
        offcurvesAux[1] = offcurves[2]
        offcurvesAux[2] = oc_new
        offcurves = offcurvesAux

        # Update 'L0' and 'D0'
        L0 = newL0
        D0 = newD0

        # Update penalties
        penalAll += np.array(penal)

        # Save profit
        profitAll.append(profit0)

        # Plot  curves
        if log:
            plotVF(2, CC, FF, PARAMS)
            plotOfferCurves(offerCurves=oc_new, hour=h, data=Bm_dict1)

    elif h >21:
        # Three-stage 
        oc_new, newL0, newD0, profit_new, profit0, penal = runThreeStageBorder(hour=h, params=PARAMS, data=Bm_dict1, probMat=Prob_ay1, 
                                                           offerCurves=offcurves, L0=L0,D0=D0, CC=CC, useVF=False, log=log)

        # Update offer curves
        offcurvesAux = {}
        offcurvesAux[0] = offcurves[1]
        offcurvesAux[1] = offcurves[2]
        offcurvesAux[2] = oc_new
        offcurves = offcurvesAux

        # Update 'L0' and 'D0'
        L0 = newL0
        D0 = newD0

        # Update penalties
        penalAll += np.array(penal)

        # Save profit
        profitAll.append(profit0)

        # Plot  curves
        #if log:
        #    plotOfferCurves(offerCurves=oc_new, hour=h, data=Bm_dict1)

        
    

        
```

No look-ahead version:

```python
useVF=False
log=True

offcurves = buildOfferCurves()
L0 = PARAMS['L0']
D0 = PARAMS['D0']
profitAll2 = []
penalAll2 = np.array([0,0,0,0])


for h in range(1,25):
    if h <= 21:
        # Three-stage 
        oc_new, newL0, newD0, profit_new, profit0, penal = runThreeStage(hour=h, params=PARAMS, data=Bm_dict1, probMat=Prob_ay1, 
                                                           offerCurves=offcurves, L0=L0,D0=D0, CC=CC, useVF=useVF, log=log)

        # Update offer curves
        offcurvesAux = {}
        offcurvesAux[0] = offcurves[1]
        offcurvesAux[1] = offcurves[2]
        offcurvesAux[2] = oc_new
        offcurves = offcurvesAux

        # Update 'L0' and 'D0'
        L0 = newL0
        D0 = newD0

        # Update penalties
        penalAll2 += np.array(penal)

        # Save profit
        profitAll2.append(profit0)

        # Plot  curves
        if log: plotOfferCurves(offerCurves=oc_new, hour=h, data=Bm_dict1)
    
    elif h > 21:
        # Three-stage 
        oc_new, newL0, newD0, profit_new, profit0, penal = runThreeStageBorder(hour=h, params=PARAMS, data=Bm_dict1, probMat=Prob_ay1, 
                                                           offerCurves=offcurves, L0=L0,D0=D0, CC=CC, useVF=useVF, log=log)

        # Update offer curves
        offcurvesAux = {}
        offcurvesAux[0] = offcurves[1]
        offcurvesAux[1] = offcurves[2]
        offcurvesAux[2] = oc_new
        offcurves = offcurvesAux

        # Update 'L0' and 'D0'
        L0 = newL0
        D0 = newD0

        # Update penalties
        penalAll2 += np.array(penal)

        # Save profit
        profitAll2.append(profit0)
```

Calculate with look-ahead and no look-ahead total profit:

```python
withVF = []
withoutVF = []

print("Profit Comparison:")
for row in range(0,len(profitAll)):
    print("time: {} -- rho_up={:.1f}/{:.1f}, rho_dw={:.1f}/{:.1f}, c={:.1f}/{:.1f}".format(
          row+1,profitAll[row][1],profitAll2[row][1],
                profitAll[row][2],profitAll2[row][2],
                profitAll[row][3],profitAll2[row][3] ))
    
    withVF.append( profitAll[row][0] + profitAll[row][1] - profitAll[row][2] - profitAll[row][3])
    withoutVF.append( profitAll2[row][0] + profitAll2[row][1] - profitAll2[row][2] - profitAll2[row][3])

print("")
print("Profit withVF: ", sum(withVF))
print("Profit withoutVF: ", sum(withoutVF))
```

```python
print("Penalties LA:    s0={} s1={} s2={} s3={}".format(penalAll[0],penalAll[1],penalAll[2],penalAll[3]))
print("Penalties no LA: s0={} s1={} s2={} s3={}".format(penalAll2[0],penalAll2[1],penalAll2[2],penalAll2[3]))
```
